```python
import pandas as pd
import openpyxl
```

```python
def sumproduct (l1,l2):
    a = 0
    k = len(l1)
    for i in range(k):
        a = a + l1[i]*l2[i]
    return a
```

```python
f_name = 'New_Excel.xlsx'
```

```python
wb = openpyxl.load_workbook(f_name)
```

Usar diretamente o engine openpyxl para abrir um excel, ao invés do pandas diretamente como no exemplo anterior (Notas Para Calcular)

```python
ws_notas = wb['Notas'] #criar novas sheets
ws_pesos = wb['Pesos']
```

```python
data_notas =  ws_notas.values #criar varíavel que vai conter os valores da worksheet notas
colunas = next(data_notas) #o next vai registrando as próximas linhas da sheet. Neste caso estamos buscando a primeira linha, que são os nomes das colunas
df_notas = pd.DataFrame(data_notas, columns=colunas) #criar um data frame
```

```python
df_notas
```

```python
data_pesos = ws_pesos.values
nomes_colunas = next(data_pesos)
df_pesos = pd.DataFrame(data_pesos, columns=nomes_colunas)
```

```python
df_pesos
```

```python
lista_pesos =list(df_pesos.loc[0, 'Teste 1': 'Lab 4'])

for i in range(len(df_notas)):
    lista_notas = list(df_notas.loc[i, 'Teste 1':'Lab 4'])
    df_notas.loc[i,'Final'] = sumproduct(lista_notas,lista_pesos)
```

```python
df_notas.head()
```

```python
k = len(df_notas)
for i in range (k):
    x = df_notas.loc[i,'Final']
    ws_notas[f'K{i+2}'] = x #i+2 pq queremos começar a escrever a partir da linha 2
```

```python
wb.save(f_name)
wb.close()
```
